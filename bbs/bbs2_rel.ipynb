```python
df_m = df[df['Geschlecht']=='männlich']
df_w = df[df['Geschlecht']=='weiblich']
df_a = df[df['Geschlecht']=='andere Identität']

df_1 = df[df['Klassenstufe']=='1. Klasse']
df_2 = df[df['Klassenstufe']=='2. Klasse']
df_3 = df[df['Klassenstufe']=='3. Klasse']

df_sg = df[df['Einschätzung_Leistung']=='sehr gut']
df_g = df[df['Einschätzung_Leistung']=='gut']
df_d = df[df['Einschätzung_Leistung']=='durchschnittlich']
df_k = df[df['Einschätzung_Leistung']=='knapp']

df_bc = df[df['Schwerpunktfach']=='Biologie/Chemie']
df_pa = df[df['Schwerpunktfach']=='Physik/Anwendungen der Mathematik']

df_ug = df[df['Schulische Herkunft']=='UG']
df_sek = df[df['Schulische Herkunft']=='Sek']
```

```python
subjects = {'Bildnerisches Gestalten': 'BG', 'BildnerischesGestalten': 'BG', 'Biologie': 'B', 'Chemie': 'C', 'Deutsch': 'D', 
            'Englisch': 'E', 'Französisch/Italienisch': 'F/I', 'Französisch / Italienisch': 'F/I','Geografie': 'Gg', 'Geschichte': 'G', 'Informatik': 'Inf', 
            'Mathematik': 'M', 'Physik': 'P', 'Musik': 'Mu',
            'Schulärztin': 'Schulärztin', 'Schulsozialarbeiter': 'Schulsozialarbeiter', 
            'Beratungslehrpersonen': 'Beratungslehrpersonen', 'Klassenlehrperson': 'Klassenlehrperson'}


df_groups = [[df], [df_m, df_w, df_a], [df_1, df_2, df_3], [df_bc, df_pa], [df_sg, df_g, df_d, df_k], [df_ug, df_sek]]
label_groups = [['alle'], ['nur männlich', 'nur weiblich', 'nur andere Identität'], ['1. Klasse', '2. Klasse', '3. Klasse'], 
                ['Biologie/Chemie', 'Physik/AM'], 
                ['sehr gut', 'gut', 'durchschnittlich', 'knapp'], ['Herkunft UG', 'Herkunft Sek']]
folders = ['alle/', 'geschlecht/', 'klasse/', 'schwerpunkt/', 'leistung/', 'herkunft/']
color_groups = [['Black'], ['CornflowerBlue', 'LightPink', 'DarkSalmon'], ['AntiqueWhite', 'BurlyWood', 'DarkGoldenRod'], 
                ['DarkSlateBlue', 'FireBrick'], 
                ['LightGreen', 'MediumSeaGreen', 'LightSlateGrey', 'PaleVioletRed'], ['MediumTurquoise', 'PaleVioletRed']]
titles = ['Alle Antworten (rel.)', 'Verteilung nach Geschlecht (rel.)', 'Verteilung nach Klassenstufe (rel.)', 'Verteilung nach Schwerpunktfach (rel.)', 'Verteilung nach Selbsteinschätzung (rel.)', 'Verteilung nach Herkunft (rel.)']

stack_max = 9

for df_group, labels, folder, colors, title in zip(df_groups, label_groups, folders, color_groups, titles):
    for index, row in cols.iterrows():
        if (row['col_include'] == 1.0) | (row['col_include'] == 3.0):
            column = row['col_name']
            suptitle = row['col_description']
            fn = sanitize_filename(column)
            path = 'figures/'+folder+fn+'_rel'+'.png'

            if (row['col_diag'] == 'bar') | (row['col_diag'] == 'pie'):
                order = row['labels']

                data = pd.concat([(group[column].value_counts()/group[column].value_counts().sum()).reindex(order) for group in df_group], axis=1)
                data.columns = labels
                
                fig = plt.figure(figsize=(8, 5), dpi=150)
                data.plot.bar(color=colors, ax=plt.gca())
                plt.suptitle(suptitle)
                plt.title(title)
                plt.grid(False)
                plt.tight_layout()
                plt.savefig(path)
                plt.close()


            if row['col_diag'] == 'bar_exp':
                order = row['labels']

                data = pd.concat([(group[column].explode(column).map(subjects).value_counts()/group[column].value_counts().sum()).reindex(order) for group in df_group], axis=1)
                data.columns = labels
                
                fig = plt.figure(figsize=(8, 5), dpi=150)
                data.plot.bar(color=colors, ax=plt.gca())
                plt.suptitle(suptitle)
                plt.title(title)
                plt.grid(False)
                plt.tight_layout()
                plt.savefig(path)
                plt.close()


            if row['col_diag'] == 'hist':
                
                bins = 5
                
                N = len(df_group)
                w = 0.6/N
                low = 0.5 - (N-1)/2 * w

                intervals = []
                for i in range(N):
                    intervals.append((low+i*w, low+5+i*w))

                fig = plt.figure(figsize=(8, 5), dpi=150)
                for group, color, label, r in zip(df_group, colors, labels, intervals):
                    group[column].hist(bins=bins, range=r, rwidth=0.9*w, density=True, color=color, label=label)

                plt.suptitle(suptitle)
                plt.title(title)
                # plt.xticks(range(1,7))
                plt.grid(False)
                plt.legend()
                plt.tight_layout()
                plt.savefig(path)
                plt.close()

    # mean = data.mean()
    # title = title + f' (⌀ = {mean:.2f})'
```
